```python
import pandas as pd
import numpy as np

# Define Matrix A
A = np.array([[4276]])

# Define Matrix B (assuming similar structure)
B = np.array([[5384]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Calculate and print the rank of Matrix A
print("Rank of A:", np.linalg.matrix_rank(A))
```
