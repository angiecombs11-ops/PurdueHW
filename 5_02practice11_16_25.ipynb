```python
import pandas as pd
import seaborn as sns
# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('housing_data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
sns.stripplot(x='Neighborhood', y='SalePrice', data=df, jitter=0.2, size=3) # Added jitter for better spread
plt.title('SalePrice Distribution by Neighborhood (Stripplot)')
plt.xlabel('Neighborhood')
plt.ylabel('SalePrice')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

This stripplot visualizes the distribution of 'SalePrice' for each 'Neighborhood'.

*   Similar to a swarm plot, each point represents an individual house's sale price within its respective neighborhood.
*   The `jitter` parameter adds a small amount of random noise to the x-axis, helping to spread out overlapping points horizontally so that individual data points can be seen, especially when there are many points at the same y-value.
*   This plot is effective for showing the raw distribution of data points for each category and is a good alternative when swarm plots become too dense.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
sns.swarmplot(x='Neighborhood', y='SalePrice', data=df, size=3) # Decreased marker size
plt.title('SalePrice Distribution by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('SalePrice')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

This swarm plot shows the distribution of 'SalePrice' for each 'Neighborhood'.

*   Each point represents an individual house's sale price within its respective neighborhood.
*   The points are adjusted so they don't overlap, providing a clearer view of the density and spread of prices.
*   This plot is useful for comparing the typical 'SalePrice' and its variability across different neighborhoods, and for identifying potential outliers within each category.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the joint plot
sns.jointplot(x='LotArea', y='SalePrice', data=df, kind='reg') # 'kind' can be 'scatter', 'kde', 'hist', 'hex', 'reg'
plt.suptitle('Joint Plot of LotArea vs. SalePrice', y=1.02) # Add a title above the plot
plt.show()
```

This joint plot visualizes the relationship between 'LotArea' and 'SalePrice'.

*   **Scatter plot (center):** Shows the direct relationship between 'LotArea' and 'SalePrice'. Each point represents a data entry, and you can observe trends, clusters, or outliers.
*   **Histograms/KDEs (margins):** Display the individual distributions of 'LotArea' (top) and 'SalePrice' (right). These help understand the spread and frequency of values for each variable independently.
*   **Regression line (optional, if `kind='reg'`):** Shows a linear regression fit, indicating the general trend between the two variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

This heatmap visualizes the correlation matrix of the numerical features.

*   **Colors:** Red colors indicate positive correlation (as one variable increases, the other tends to increase). Blue colors indicate negative correlation (as one variable increases, the other tends to decrease). White or light colors indicate a weak or no correlation.
*   **Numbers:** The numbers in each cell represent the Pearson correlation coefficient, ranging from -1 to 1. A value closer to 1 or -1 indicates a stronger linear relationship.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the numerical variables for the pair plot
selected_columns = ['LotArea', 'YearBuilt', 'SalePrice']

# Create the pair plot
sns.pairplot(df[selected_columns])
plt.suptitle('Pair Plot of LotArea, YearBuilt, and SalePrice', y=1.02) # Add a title above all subplots
plt.show()
```

This pair plot visualizes the relationships between 'LotArea', 'YearBuilt', and 'SalePrice'.

*   **Diagonal plots (histograms/KDEs):** Show the distribution of each individual variable.
*   **Off-diagonal plots (scatter plots):** Show the pairwise relationship between different variables. For instance, you can observe how 'SalePrice' changes with 'LotArea' or 'YearBuilt', and vice-versa. Patterns in these scatter plots can indicate correlations or trends.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.violinplot(y=df['SalePrice'])
plt.title('Distribution of SalePrice')
plt.ylabel('SalePrice')
plt.show()
```

This violin plot illustrates the distribution of 'SalePrice'. The wider sections of the violin indicate a higher probability density (more data points) at those price levels, while the thinner sections indicate lower density. The box plot inside shows the interquartile range (IQR), median, and potential outliers.
