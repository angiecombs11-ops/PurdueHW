```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the items.csv dataset
df_items = pd.read_csv('items.csv')
print("\n--- items.csv ---")
print("Shape:", df_items.shape)
df_items.info()
print("\nDescriptive Statistics:")
print(df_items.describe())

# Check for outliers in numerical columns for items.csv
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df_items['kcal'])
plt.title('Box plot of Kcal (items.csv)')
plt.subplot(1, 2, 2)
sns.boxplot(y=df_items['cost'])
plt.title('Box plot of Cost (items.csv)')
plt.tight_layout()
plt.show()

# Load the resturants.csv dataset
df_restaurants = pd.read_csv('resturants.csv')
print("\n--- resturants.csv ---")
print("Shape:", df_restaurants.shape)
df_restaurants.info()
print("\nDescriptive Statistics:")
print(df_restaurants.describe())
# Note: resturants.csv does not have numerical columns suitable for box plot outlier detection

# Load the sales.csv dataset
df_sales = pd.read_csv('sales.csv')
print("\n--- sales.csv ---")
print("Shape:", df_sales.shape)
df_sales.info()
print("\nDescriptive Statistics:")
print(df_sales.describe())

# Check for outliers in numerical columns for sales.csv
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df_sales['price'])
plt.title('Box plot of Price (sales.csv)')
plt.subplot(1, 2, 2)
sns.boxplot(y=df_sales['item_count'])
plt.title('Box plot of Item Count (sales.csv)')
plt.tight_layout()
plt.show()
```

```python
print("--- Outlier Ranges for df_sales ---")

# Calculate outlier ranges for 'price' column
Q1_price = df_sales['price'].quantile(0.25)
Q3_price = df_sales['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
lower_bound_price = Q1_price - 1.5 * IQR_price
upper_bound_price = Q3_price + 1.5 * IQR_price

print(f"\nPrice Column:\n  Q1: {Q1_price:.2f}\n  Q3: {Q3_price:.2f}\n  IQR: {IQR_price:.2f}")
print(f"  Lower Outlier Range (below): {lower_bound_price:.2f}")
print(f"  Upper Outlier Range (above): {upper_bound_price:.2f}")

# Calculate outlier ranges for 'item_count' column
Q1_item_count = df_sales['item_count'].quantile(0.25)
Q3_item_count = df_sales['item_count'].quantile(0.75)
IQR_item_count = Q3_item_count - Q1_item_count
lower_bound_item_count = Q1_item_count - 1.5 * IQR_item_count
upper_bound_item_count = Q3_item_count + 1.5 * IQR_item_count

print(f"\nItem Count Column:\n  Q1: {Q1_item_count:.2f}\n  Q3: {Q3_item_count:.2f}\n  IQR: {IQR_item_count:.2f}")
print(f"  Lower Outlier Range (below): {lower_bound_item_count:.2f}")
print(f"  Upper Outlier Range (above): {upper_bound_item_count:.2f}")
```

```python
print(f"Shape of df_sales before filtering: {df_sales.shape}")

# Filter out rows where 'price' is greater than 39.05 or less than 1.00
df_sales_filtered = df_sales[~((df_sales['price'] > 39.05) | (df_sales['price'] < 1.00))]

print(f"Shape of df_sales after filtering price outliers: {df_sales_filtered.shape}")
```

```python
print("Distinct Item IDs after filtering price outliers:")
distinct_item_ids = df_sales_filtered['item_id'].unique()
print(distinct_item_ids)
print(f"Number of distinct item IDs: {len(distinct_item_ids)}")
```

```python
print("--- Outlier Ranges for df_items (kcal and cost) ---")

# Calculate outlier ranges for 'kcal' column
Q1_kcal = df_items['kcal'].quantile(0.25)
Q3_kcal = df_items['kcal'].quantile(0.75)
IQR_kcal = Q3_kcal - Q1_kcal
lower_bound_kcal = Q1_kcal - 1.5 * IQR_kcal
upper_bound_kcal = Q3_kcal + 1.5 * IQR_kcal

print(f"\nKcal Column:\n  Q1: {Q1_kcal:.2f}\n  Q3: {Q3_kcal:.2f}\n  IQR: {IQR_kcal:.2f}")
print(f"  Lower Outlier Range (below): {lower_bound_kcal:.2f}")
print(f"  Upper Outlier Range (above): {upper_bound_kcal:.2f}")

# Calculate outlier ranges for 'cost' column
Q1_cost = df_items['cost'].quantile(0.25)
Q3_cost = df_items['cost'].quantile(0.75)
IQR_cost = Q3_cost - Q1_cost
lower_bound_cost = Q1_cost - 1.5 * IQR_cost
upper_bound_cost = Q3_cost + 1.5 * IQR_cost

print(f"\nCost Column:\n  Q1: {Q1_cost:.2f}\n  Q3: {Q3_cost:.2f}\n  IQR: {IQR_cost:.2f}")
print(f"  Lower Outlier Range (below): {lower_bound_cost:.2f}")
print(f"  Upper Outlier Range (above): {upper_bound_cost:.2f}")
```

```python
# Merge df_sales_filtered with df_items
# Rename 'name' column in df_items to 'item_name' before merging to avoid conflict
merged_df = pd.merge(df_sales_filtered, df_items.rename(columns={'name': 'item_name'}),
                     left_on='item_id', right_on='id', how='left')

# Drop the redundant 'id' column from df_items after merge
merged_df.drop(columns=['id'], inplace=True)

# Merge the result with df_restaurants
# Rename 'name' column in df_restaurants to 'store_name' before merging to avoid conflict
final_df = pd.merge(merged_df, df_restaurants.rename(columns={'name': 'store_name'}),
                    left_on='store_id', right_on='id', how='left')

# Drop the redundant 'id' column from df_restaurants after merge
final_df.drop(columns=['id'], inplace=True)

print("First 5 rows of the merged dataset:")
display(final_df.head())

print("\nInformation about the merged dataset:")
final_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Ensure pandas is imported for pd.api.types

# Ensure 'date' is datetime type
if not pd.api.types.is_datetime64_any_dtype(final_df['date']):
    final_df['date'] = pd.to_datetime(final_df['date'])

# Ensure 'total_sales' column is present
if 'total_sales' not in final_df.columns:
    final_df['total_sales'] = final_df['price'] * final_df['item_count']

# Ensure 'day_of_week' column is present
if 'day_of_week' not in final_df.columns:
    final_df['day_of_week'] = final_df['date'].dt.dayofweek

# Calculate total sales for each specific date and day of the week
daily_sales_by_day_of_week = final_df.groupby(['date', 'day_of_week'])['total_sales'].sum().reset_index()

# Then calculate the average of these daily total sales for each day of the week
average_sales_by_day_of_week = daily_sales_by_day_of_week.groupby('day_of_week')['total_sales'].mean().reset_index()

# Map day numbers to names for better readability
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
average_sales_by_day_of_week['day_name'] = average_sales_by_day_of_week['day_of_week'].map(lambda x: day_names[x])

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='day_name', y='total_sales', data=average_sales_by_day_of_week, palette='viridis', hue='day_name', legend=False)
plt.title('Average Daily Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Daily Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Average Daily Sales by Quarter Over Years

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'date' is datetime type
if not pd.api.types.is_datetime64_any_dtype(final_df['date']):
    final_df['date'] = pd.to_datetime(final_df['date'])

# Ensure 'total_sales' column is present
if 'total_sales' not in final_df.columns:
    final_df['total_sales'] = final_df['price'] * final_df['item_count']

# Ensure 'quarter' and 'year' columns are present
if 'quarter' not in final_df.columns:
    final_df['quarter'] = final_df['date'].dt.quarter
if 'year' not in final_df.columns:
    final_df['year'] = final_df['date'].dt.year

# Calculate average daily sales for each quarter per year
# First, sum total_sales by date, year, and quarter to get daily total sales
daily_sales_per_year_quarter = final_df.groupby([final_df['date'].dt.date, 'year', 'quarter'])['total_sales'].sum().reset_index()

# Then, calculate the mean of these daily total sales for each year and quarter
average_sales_by_year_quarter = daily_sales_per_year_quarter.groupby(['year', 'quarter'])['total_sales'].mean().reset_index()

print("Average Daily Sales by Year and Quarter:")
display(average_sales_by_year_quarter)

plt.figure(figsize=(12, 7))
sns.lineplot(x='quarter', y='total_sales', hue='year', data=average_sales_by_year_quarter, marker='o', palette='viridis')
plt.title('Average Daily Sales by Quarter Over Years')
plt.xlabel('Quarter')
plt.ylabel('Average Daily Sales')
plt.xticks(ticks=[1, 2, 3, 4], labels=['Q1', 'Q2', 'Q3', 'Q4'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

The plot 'Average Daily Sales by Quarter Over Years' shows a clear upward trend in average daily sales across all quarters from 2019 to 2021. This indicates consistent growth year-over-year. For each year, sales consistently peak in the second and third quarters (Q2 and Q3), reinforcing the mid-year sales strength observed in the overall quarterly analysis. The first and fourth quarters (Q1 and Q4) consistently show lower average daily sales, though these also demonstrate growth over the years.

### Preparing Data for Machine Learning Models

```python
from datetime import timedelta

# Sort the DataFrame by date to ensure proper splitting
final_df_sorted = final_df.sort_values('date')

# Define the target variable
X = final_df_sorted.drop(['date', 'total_sales', 'item_name', 'restaurant_name'], axis=1)
y = final_df_sorted['total_sales']

# Identify the split point for the last six months
last_date = final_df_sorted['date'].max()
six_months_ago = last_date - pd.DateOffset(months=6)

# Split data into training and testing sets
train_df = final_df_sorted[final_df_sorted['date'] < six_months_ago]
test_df = final_df_sorted[final_df_sorted['date'] >= six_months_ago]

X_train = train_df.drop(['date', 'total_sales', 'item_name', 'restaurant_name'], axis=1)
X_test = test_df.drop(['date', 'total_sales', 'item_name', 'restaurant_name'], axis=1)
y_train = train_df['total_sales']
y_test = test_df['total_sales']

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
print(f"Test data starts from: {test_df['date'].min()}")
print(f"Test data ends on: {test_df['date'].max()}")
```

### Model Building and Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize models
lr_model = LinearRegression()
rf_model = RandomForestRegressor(random_state=42)
xgb_model = XGBRegressor(random_state=42)

models = {
    'Linear Regression': lr_model,
    'Random Forest': rf_model,
    'XGBoost': xgb_model
}

rmse_results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    rmse_results[name] = rmse
    print(f"{name} RMSE: {rmse:.2f}")

# Compare models
best_model_name = min(rmse_results, key=rmse_results.get)
best_rmse = rmse_results[best_model_name]

print(f"\nBest performing model: {best_model_name} with RMSE: {best_rmse:.2f}")
```

### Forecasting for the Next Year
