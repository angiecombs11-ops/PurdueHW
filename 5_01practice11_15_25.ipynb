```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('housing_data.csv')
# Generate a line plot
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.scatter(df['LotArea'], df['OverallQual'], df['SalePrice'], c='skyblue', s=50, alpha=0.6)

# Set labels and title
ax.set_xlabel('Lot Area')
ax.set_ylabel('Overall Quality')
ax.set_zlabel('Sale Price')
ax.set_title('3D Scatter Plot of LotArea, OverallQual, and SalePrice')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the proportion of houses in each MSZoning category
mszoning_counts = df['MSZoning'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(mszoning_counts, labels=mszoning_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.viridis.colors)
plt.title('Proportion of Houses by MSZoning Category')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
summary_stats_by_neighborhood = df.groupby('Neighborhood')['SalePrice'].describe()
display(summary_stats_by_neighborhood)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.boxplot(x='Neighborhood', y='SalePrice', data=df, palette='viridis', hue='Neighborhood', legend=False)
plt.title('Distribution of SalePrice by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Sale Price')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.countplot(y='Neighborhood', hue='Neighborhood', data=df, order = df['Neighborhood'].value_counts().index, palette='viridis', legend=False)
plt.title('Count of Houses by Neighborhood')
plt.xlabel('Number of Houses')
plt.ylabel('Neighborhood')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='LotArea', y='SalePrice', data=df, alpha=0.6)
plt.title('Relationship Between Lot Area and Sale Price')
plt.xlabel('Lot Area (square feet)')
plt.ylabel('Sale Price')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'YrSold' and calculate the mean of 'SalePrice'
year_price_trend = df.groupby('YrSold')['SalePrice'].mean().reset_index()

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='YrSold', y='SalePrice', data=year_price_trend)
plt.title('Average House Prices Over the Years')
plt.xlabel('Year Sold')
plt.ylabel('Average Sale Price')
plt.grid(True)
plt.xticks(year_price_trend['YrSold'].unique())
plt.show()
```
