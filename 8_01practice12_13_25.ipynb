```python
import pandas as pd
from scipy import stats

# Load the dataset
data = pd.read_csv('cutlets.csv')

# Separate the diameter sizes of the cutlets
Unit_A = data['Unit A']
Unit_B = data['Unit B']
```

```python
# Perform the independant two-sample t-test
t_stat, p_value = stats.ttest_ind(Unit_A, Unit_B, equal_var=True)
print("P_value: {p_val}")
```

# Intrepretation

```python
# Intrepretation
alpha = .05
if p_value < alpha:
  print("Reject the null hypothesis - There is a significant difference between Unit A and Unit B cutlet size")
else:
  print("Fail to reject the null hypothesis - no significant difference between Unit A and Unit B cutlet size")
```
